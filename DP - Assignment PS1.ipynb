```python
using System;
public class DPAssignment
{
    public static void Main()
    {
        int[] ar1 = new int[5];  //holds all the value we give as an input.
        int[] ar2 = new int[5];  //holds all the value of even integer.
        int[] ar3 = new int[5];  //holds all the value of odd integer.
        int a = 0;
        int b = 0;
        int c = 0;

        for (a = 0; a < 5; a++)
        {
            Console.Write($"Input Number {a} : ");  // taking input from user
            ar1[a] = Convert.ToInt32(Console.ReadLine());
        }
        for (a = 0; a < 5; a++)                //for loop will give all the values one by one
        {
            if (ar1[a] % 2 == 0)              // If loop check whether the value is devided by 2 or not. 
            {
                ar2[b] = ar1[a];             //If it divided then the value will go into bucket 2
                b++;
            }
            else                             //else the value will go to bucket 3
            {
                ar3[c] = ar1[a];
                c++;
            }
        }
                                             //seperate integers goes into different bucket and will be printed.
        Console.WriteLine("The Even numbers are : ");
        for (a = 0; a < b; a++)
        {
            Console.Write($"{ar2[a]}, ");
        }

        Console.WriteLine("\nThe Odd numbers are : ");
        for (a = 0; a < c; a++)
        {
            Console.Write($"{ar3[a]}, ");
        }
    }
}
```
